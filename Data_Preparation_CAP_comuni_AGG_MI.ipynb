```python
import pandas as pd
import numpy as np
```

```python
zonizzazione = pd.read_csv('Data/zonizzazione/zonizzazione_ABCD.csv')
zonizzazione
```

```python
AGG_MI = zonizzazione[zonizzazione['Zona'] == 'AGG MI']
AGG_MI
```

```python
listacomuni = pd.read_csv('Data/zonizzazione/listacomuni.csv', sep = ';', encoding = 'latin')
listacomuni
```

```python
#comuni_AGG_MI = pd.merge(AGG_MI[['Codice Istat', 'Provincia']], listacomuni[['Comune', 'CAP', 'Istat']], left_on = 'Codice Istat', right_on = 'Istat')
comuni_AGG_MI = pd.merge(AGG_MI[['Provincia']], listacomuni[['Comune', 'CAP', 'Istat']], left_on = AGG_MI['Comune'].str.lower(), right_on = listacomuni['Comune'].str.lower())
comuni_AGG_MI
```

```python
ix =['x' in x for x in comuni_AGG_MI['CAP']]
milano = pd.DataFrame(np.repeat(comuni_AGG_MI[ix].values, 20162-20121+1, axis=0))
milano['cap'] = range(20121, 20163)
milano = milano[[0, 1, 2, 'cap', 4]]
milano.set_axis(['Codice Istat', 'Provincia', 'Comune', 'CAP', 'Istat'], axis = 1, inplace= True)
milano
```

```python
CAP_AGG_MI = comuni_AGG_MI[['x' not in x for x in comuni_AGG_MI['CAP']]]
CAP_AGG_MI = pd.concat([CAP_AGG_MI, milano], ignore_index = True)
CAP_AGG_MI.sort_values('Comune', inplace = True, ignore_index = True)
CAP_AGG_MI.drop(['key_0', 'Codice Istat', 'Istat'], axis = 1, inplace = True)
CAP_AGG_MI
```

```python
CAP_AGG_MI.to_csv('Data/data_viz/CAP_AGG_MI.csv')
```
